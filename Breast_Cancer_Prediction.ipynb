# Breast Cancer Prediction:

Health is an essential aspect of everyone’s life. Breast cancer is found in the body of male or female when the cells in the breast begin to grow out of control. These cells usually form a tumor and can be felt as a lump or could be seen on an x-ray. Cancer can be distinguished as benign, or either can be malignant (cancer).


```python
# Step 2 : import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2 : import data
df=pd.read_csv('/content/Breast_Cancer.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()   # checks null values
```

```python
df.describe() #understand data
```

```python
df.columns
```

```python
# Step 3 : define target (y) and features (X)
x=df.drop(['id','diagnosis'],axis=1)
y=df['diagnosis']
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=200)
```

```python
# check shape of train and test sample
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
# Step 6 : train or fit mod
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred=model.predict(x_test)
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
